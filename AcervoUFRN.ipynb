```python
# excluir coluna registro_sistema porque não agrega vlor para a análise
Dados = Dados.drop('registro_sistema', axis=1)
```

```python
# incluindo coluna explicando localização
def localizacao(linha):
  #CDU = []
  if linha <= 99:
    return 'Generalidades. Ciência e conhecimento'
  elif  linha <= 199:
    return 'Filosofia e psicologia'
  elif  linha <= 299:
    return 'Religião'
  elif  linha <= 399:
    return 'Ciências sociais'
  elif  linha <= 499:
    return 'Classe vaga. Provisoriamente não ocupada'
  elif  linha <= 599:
    return 'Matemática e ciências naturais'
  elif  linha <= 699:
    return 'Ciências aplicadas'
  elif  linha <= 799:
    return 'Belas artes'
  elif  linha <= 899:
    return 'Linguagem. Língua. Linguística'
  elif  linha <= 999:
    return 'Geografia. Biografia. História'
  else:
    return 0
```

```python
Dados['CDU'] = Dados["localizacao"].map(localizacao)
```

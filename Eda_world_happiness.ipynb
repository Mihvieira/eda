## Visão geral do Dataset
O Relatório Mundial da felicidade (World Happiness Report) foi lançado pelas Nações Unidas em comemoração ao dia internacional da felicidade em 20/03.
Ele traz dados históricos sobre o estado global da felicidade entre os anos de 2013 a 2017, classificando 155 países pelos níveis deste sentimento.

Obs. Distopia é um país hipotético que contem valores iguais a média nacional mais baixa do mundo para cada um dos 6 fatores

```python
y = acima_média['Region']
x = acima_média['Happiness Score']

fig, ax = plt.subplots()

ax.stem(y,x)
plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.show()
```

```python
# visualizando a pontuação de felicidade por regiões e comparando com a média
y = df2015['Region']
x = df2015['Happiness Score']

markerline, stemlines, baseline = plt.stem(
    y, x, linefmt='--', markerfmt='o', bottom= média_2015)
markerline.set_markerfacecolor('none')

plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.show()
```

```python
# Plot para Freedom
region = df2015['Region']
f = df2015['Freedom']
Freedom_2015 = df2015['Freedom'].mean()

# Plot para Trust
t = df2015['Trust (Government Corruption)']
Trust_2015 = df2015['Trust (Government Corruption)'].mean()

# Plot para Generosity
g = df2015['Generosity']
Generosity_2015 = df2015['Generosity'].mean()

# Plot para Renda per capita
e = df2015['Economy (GDP per Capita)']
gdp_2015 = df2015['Economy (GDP per Capita)'].mean()

# tamanho da figura
fig, ((ax, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True, sharey=True, figsize = (10,12))
plt.tight_layout(h_pad=2)

fig.suptitle('Region')
plt.subplots_adjust(top=0.90)

ax.stem(region, f, linefmt=(':', 'slateblue'), markerfmt='D', bottom= Freedom_2015, orientation=  'horizontal')
ax.set_title('Freedom', )

ax2.stem(region, t, linefmt=(':', 'darkslateblue'), markerfmt='D', bottom= Trust_2015, orientation=  'horizontal')
#ax2.axis('equal')
ax2.set_title('Trust')

ax3.stem(region, g, linefmt=(':', 'mediumslateblue'), markerfmt='D', bottom= Generosity_2015, orientation=  'horizontal')
ax3.set_title('Generosity')

ax4.stem(region, e, linefmt=(':', 'mediumpurple'), markerfmt='D', bottom= gdp_2015, orientation=  'horizontal')
ax4.set_title('GDP per capita')

plt.show()
```
